## 3. ***Data Wrangling***

### Data Wrangling Code

````python
#HANDLE MISSING VALUES

# Calculate and display the number of missing values
print("Missing values before handling:")
display(df.isnull().sum())

# Identify columns with missing values (already done in previous cells, but good to reconfirm)
# Columns with missing values: 'Rating', 'Type', 'Content Rating', 'Current Ver', 'Android Ver'

# Drop rows with missing values in 'Type', 'Content Rating', 'Current Ver', 'Android Ver'
df.dropna(subset=['Type', 'Content Rating', 'Current Ver', 'Android Ver'], inplace=True)

# Fill missing values in 'Rating' with the mean
# Calculate the mean of the 'Rating' column, excluding NaN values
mean_rating = df['Rating'].mean()
df['Rating'].fillna(mean_rating, inplace=True)

# Display the number of missing values after handling
print("\nMissing values after handling:")
display(df.isnull().sum())




````

````python
#HANDLE DUPLICATES VALUES

# Check for and count the number of duplicate rows
print("Number of duplicate rows before removal:", df.duplicated().sum())

# Remove duplicate rows
df.drop_duplicates(inplace=True)

# Verify that duplicate rows have been removed
print("Number of duplicate rows after removal:", df.duplicated().sum())
````

````python
#HANDLE OUTLIERS

# Identify the row with Rating = 19.0
outlier_row = df[df['Rating'] == 19.0]
print("Row with Rating = 19.0:")
display(outlier_row)

# Remove the row with Rating = 19.0
df = df[df['Rating'] != 19.0]

# Verify that the outlier has been removed
print("\nMaximum rating after removing outlier:", df['Rating'].max())
````

````python
#CLEAN 'SIZE' COLUMNS

# Define a function to clean the 'Size' column
def clean_size(size):
    if isinstance(size, str):
        if 'M' in size:
            return float(size.replace('M', ''))
        elif 'k' in size:
            return float(size.replace('k', '')) / 1024  # Convert KB to MB
    return np.nan  # Handle 'Varies with device', non-numeric, or non-string values

# Apply the cleaning function to the 'Size' column and create a new 'Size_MB' column
df['Size_MB'] = df['Size'].apply(clean_size)

# Display the first few rows of the DataFrame including the new 'Size_MB' column
display(df.head())
````

````python
#CLEAN 'INSTALLS' COLUMNS

# Remove ',' and '+' from 'Installs' and convert to integer
df['Installs'] = df['Installs'].astype(str).str.replace('[+,]', '', regex=True)
df['Installs'] = pd.to_numeric(df['Installs'], errors='coerce')

# Display the data type and first few rows of the 'Installs' column
print("Data type of 'Installs' column after cleaning:", df['Installs'].dtype)
display(df['Installs'].head())
````

````python
#CLEAN 'PRICE' COLUMNS
# Ensure the 'Price' column is of string type and remove '$'
df['Price'] = df['Price'].astype(str).str.replace('$', '', regex=False)

# Convert the cleaned 'Price' column to a numerical type
df['Price'] = pd.to_numeric(df['Price'], errors='coerce')

# Display the data type and first few rows of the 'Price' column
print("Data type of 'Price' column after cleaning:", df['Price'].dtype)
display(df['Price'].head())
````

### What all manipulations have you done and insights you found?

The data wrangling involved checking and handling missing values, dropping duplicate rows, and filling missing ratings with the mean. Insights include the number of missing values in various columns, the count of duplicate rows, and confirmation that missing value handling was applied to a new DataFrame.

````python
# Chart - 14 visualization code

# Convert 'Reviews' to numerical, coercing errors
df['Reviews'] = pd.to_numeric(df['Reviews'], errors='coerce')

# Select numerical columns for correlation analysis
numerical_cols = ['Rating', 'Reviews', 'Size_MB', 'Price']
df_numerical = df[numerical_cols]

# Calculate the correlation matrix
correlation_matrix = df_numerical.corr()

# Create a heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")

# Add title
plt.title('Correlation Heatmap of Numerical Variables')

plt.show()
````

I chose the correlation heatmap because it's the most effective way to visualize the pairwise correlations between multiple numerical variables. The color intensity and annotations clearly show the strength and direction of the relationships at a glance.

'Price' has very weak negative correlations with 'Rating', 'Reviews', and 'Size_MB', suggesting that price has minimal linear relationship with these variables.
